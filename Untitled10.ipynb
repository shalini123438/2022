# null value imputation

```python
df['age']=df['age'].fillna(df['age'].mean())
```

```python
df['fare']=df['fare'].fillna(df['fare'].mean())
```

```python
df['cabin']=df['cabin'].fillna(df['cabin'].mode())
```

```python
df['embarked']=df['embarked'].fillna(df['embarked'].mode())
```

```python
df['boat']=df['boat'].fillna(df['boat'].mode(),inplace=True)
```

```python
df['body']=df['body'].fillna(df['body'].mean())
```

```python
df['home.dest']=df['home.dest'].fillna(statistics.mode(df['home.dest']))
```

```python
df.isnull().sum()  #not able to update categorical value
```

```python
df.drop(['ticket'],axis=1,inplace=True)
```

# Checking for data distribution

```python
#checking destibution of age data with the help of distplot
```

```python
sns.set_style('whitegrid')
sns.distplot(df['age'], kde = True, color ='red', bins = 30)
```

```python
#shapiro test for age data distribution
```

```python
from scipy.stats import shapiro
data_test=df['age']
stats,p=shapiro(data_test)
print(stats,p)
if p>0.05:
    print('Age is normally distributed')
else:
    print('Age is not normally distributes')  #hence with shapiro test it is proved that data is not normally distributed so
                                              #we have to convert data into nomal distribution 
```

```python
#Now will convert not distributed data into distributed data
```

```python
scaler = StandardScaler()
arr = scaler.fit_transform(df[['age']]).flatten()  #if we are using one column for standerise our data then we need to add 
#[] and .flatten()  
```

```python
#Now will check either updated data is normally distributed or not
```

```python
sns.set_style('whitegrid')
sns.distplot(arr, kde = True, color ='red', bins = 30)
```

```python
from scipy.stats import shapiro
data_test=arr
stats,p=shapiro(data_test)
print(stats,p)
if p>0.05:
    print('Age is normally distributed')
else:
    print('Age is not normally distributes')
```

# Multicolinearity (Independent->Independent)

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
```

```python
 X=df
```

```python
vif_data = pd.DataFrame()
vif_data["feature"] = X.columns
  
# calculating VIF for each feature
vif_data["VIF"] = [variance_inflation_factor(X.values, i)
                          for i in range(len(X.columns))]
  
print(vif_data)
```

# Lasso->Another model building is going on

```python
lassocv=LassoCV(cv=10,max_iter=20000000,normalize=True)
lassocv.fit(x_train,y_train)
```

```python
lasso=Lasso(alpha=lassocv.alpha_)
lasso.fit(x_train,y_train)
```

# Ridge->Another model building is going on

```python
ridgecv=RidgeCV(alphas=np.random.uniform(0,10,50),cv=10,normalize=True)
```

```python
ridgecv.fit(x_train,y_train)
```

```python
ridgecv.alpha_
```

```python
ridge_lr=Ridge(alpha=ridgecv.alpha_)
ridge_lr.fit(x_train,y_train)
```

```python
ridge_lr.score(x_test,y_test)
```
